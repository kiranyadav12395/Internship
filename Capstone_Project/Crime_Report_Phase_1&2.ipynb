Ans 2.1	Analysis of Literacy Rate vs Total Crimes.

```python
plt.figure(figsize=(22,10))
sns.lineplot(data['Literacy rate'],data['Total Crimes'])
```

Arunachal Pradesh has lowest literacy rate (65%) with a total crime figure 2286. Whereas Kerala has highest literacy rate (94%) with a total crime figure 172137. By this figure we can conclude increase in literacy rate somehow also increase the total crime figure. There is a linear relation between this value (Literacy rate & total crime).

Ans 2.2	 Analysis of the type of crime vs each state vs Literacy rate.

```python
Data=pd.read_csv('Ans_2.csv')
Data
```

```python
plt.figure(figsize=(20,20))
sns.barplot(Data['Literacy rate'],Data['MURDER'],hue=Data['State/Union Territory'])
```

When we Analyse type of crime vs each state vs Literacy rate, we have found that Uttar Pradesh have lowest literacy rate whereas large no of crime cases is registered in murder, rape, kidnaping, dacoity, robbery & theft. Whereas Lakshadweep has highest literacy rate and lower cases are registered of murder, rape, kidnaping, dacoity, robbery & theft.

Ans 2.4	 Analysis of area vs overall crime

```python
plt.figure(figsize=(20,10))
sns.lineplot(data['Area'],data['Total Crimes'])
```

Lakshadweep has lowest available area in our dataset which is 32 square meters with a total crime figure 20, whereas Rajasthan has highest available area which is 342239 square meters with a total crime figure 165622.  So, there is linear relation between available area and total crime figure. 

Ans 2.5	 Analysis of Population vs overall Crime

```python
plt.figure(figsize=(20,15))
sns.lineplot(data['Population'],data['Total Crimes'])
```

Uttar Pradesh has the highest population in our dataset which is 199812 with a total crime figure 195135, whereas Lakshadweep has lowest population which is 64 in our dataset with a crime figure 20. So, there is linear relation between population and total crim figure. 

Ans 2.6	Each state crime report. 

```python
plt.figure(figsize=(20,15))
sns.barplot(data['Total Crimes'],data['State/Union Territory'])
plt.title("State wise crime report)")
```

From the graph plot we can conclude that, highest no of crimes reported in Madhya Pradesh whereas lowest no of crimes reported in Lakshadweep. Here if we calculate total crimes per unit of population for Madhya Pradesh it is 2.8 count per person & for Lakshadweep it is 0.3 crimes per person. Literacy rate might be another reason for increase in total crimes in Madhya Pradesh state. (Literacy rate in MP: - 69% & in Lakshadweep it is 92%). So in MP we have to focus on increase in literacy rate to reduce the overall crime figure.
