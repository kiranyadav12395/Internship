```python
with open ('42_District_wise_crimes_committed_against_women_2001_2012.csv', 'r') as file:
    no_records = 0
    for row in file:
        cursor.execute("insert into crimes_scenes values (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", row.split(","))
        db.commit()
        no_records+=1
print(no_records, 'records inserted')  
```
